```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
# Load data

data = pd.read_csv("sanfrancisco_incidents_summer_2014.csv")
```

```python
## Explore Data

data.describe(include = 'all')
```

```python
data.head()
```

```python
# Exploring 

# Number of crimes per district

dataPlot = data[['PdDistrict', 'IncidntNum']].groupby(['PdDistrict']).agg(['count'])

dataPlot.plot.bar()
```

```python
# For either city, how do incidents vary by time of day? 
dataPlot1 = data
dataPlot1['hour'] = data['Time'].apply(lambda x:x.split(':')[0])
dataPlot1['IncidntNum'] = data['IncidntNum']

dataPlot1 = dataPlot1.loc[(dataPlot1['PdDistrict'] == 'SOUTHERN')]

dataPlot1 = dataPlot1[['hour', 'IncidntNum']].groupby(['hour']).agg(['count'])
dataPlot1.plot(kind='bar')
```

```python
#Which incidents are most common in the evening?
dataPlot2 = data
dataPlot2['hour'] = data['Time'].apply(lambda x:x.split(':')[0])
dataPlot2['IncidntNum'] = data['IncidntNum']
dataPlot2['hour'].fillna(0.0)
dataPlot2 = dataPlot2[dataPlot2['hour'].astype(int) > 6]

dataPlot2 = dataPlot2.loc[(dataPlot2['PdDistrict'] == 'SOUTHERN')]

dataPlot2 = dataPlot2[['hour', 'IncidntNum']].groupby(['hour']).agg(['count'])
dataPlot2.plot(kind='bar')
```

```python
# During what periods of the day are robberies most common?
dataPlot3 = data
dataPlot3['hour'] = data['Time'].apply(lambda x:x.split(':')[0]).astype(int)

dataPlot3 = dataPlot3.loc[(dataPlot3['Category'] == 'ROBBERY') & (dataPlot3['PdDistrict'] == 'SOUTHERN')]
dataPlot3[['Category', 'hour']].hist()
```

```python
# For either city, how do incidents vary by neighborhood? for the SOUTHERN district
data_neighbors =  data.loc[(data['PdDistrict'] == 'SOUTHERN')]
```

```python
# For either city, how do incidents vary month to month in the Summer 2014 dataset?
dataPlot4 = data
dataPlot4['month'] = data['Date'].apply(lambda x:x.split('/')[0])
dataPlot4['month'].fillna(0.0)
dataPlot4 = dataPlot4[(dataPlot4['month'].astype(int) > 4) & (dataPlot4['month'].astype(int) < 10)]
dataPlot4['IncidntNum'] = data['IncidntNum']
```

```python
dataPlot4 = dataPlot4.loc[(dataPlot4['PdDistrict'] == 'SOUTHERN')]
dataPlot4 = dataPlot4[['month', 'IncidntNum']].groupby(['month']).agg(['count'])
```
